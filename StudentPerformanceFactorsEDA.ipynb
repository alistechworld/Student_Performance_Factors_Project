# Bivariate Analysis

```python
df.head()
```

```python
# Selecting only numerical columns
numerical_df = df.select_dtypes(include="number")

# Creating the scatter plot matrix

sns.pairplot(numerical_df)
plt.show()
```

```python
def plot_box(df, categorical_col, numerical_col):
    # Creating the box plot
    plt.figure(figsize=(10,6))
    sns.boxplot(x=categorical_col, y = numerical_col, color = "skyblue", data = df)
    plt.title(f"Box Plot of {numerical_col} by {categorical_col}")
    plt.xlabel(categorical_col)
    plt.ylabel(numerical_col)
    plt.xticks(rotation = 45)
    plt.show()
```

```python
df.head()
```

```python
plot_box(df, "Parental_Involvement", "Exam_Score")
```
